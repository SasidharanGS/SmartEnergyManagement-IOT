```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
df = pd.read_csv("./data.csv")
df
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%Y -> %H:%M:%S', errors='coerce')

# Split the 'Timestamp' column values into separate date and time columns
df['Date'] = df['Timestamp'].dt.date
df['Time'] = df['Timestamp'].dt.time
df
```

```python
grouped = df.groupby('Date')
date_dataframes = {str(date): group for date, group in df.groupby('Date')}
date_dataframes['2023-11-05']
```

```python
unique_dates = df['Date'].unique()
unique_dates
```

```python
#Create a plot for each day with different colors
plt.figure(figsize=(12, 6))

for i, date in enumerate(unique_dates):
    specific_date_df = date_dataframes.get(str(date))  # Assuming you have already created date_dataframes
    # Extract time and power consumption
    time = specific_date_df['TimeStamp'].dt.time
    power_consumption = specific_date_df['Power (W)']
    
    # Plot the data for the day with a different color
    plt.plot(time, power_consumption, label=f'Day {i + 1}')

# Customize the plot
plt.xlabel('Time')
plt.ylabel('Power Consumption (W)')
plt.title('Power Consumption in 24 Hours for Each Day')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
In this modified code, we use specific_date_df to access each day's DataFrame from the date_dataframes dictionary and plot the power consumption for each day with different colors. The rest of the code remains the same.

Make sure to replace 'your_data.csv' with the actual path to your CSV file.
```

```python
for date, group in grouped:
    plt.figure(figsize=(10, 6))
    time_str = group['Time'].apply(lambda x: x.strftime('%H:%M:%S'))
    plt.plot(time_str, group['Power (W)'], label='Power (W)', marker='o')
    plt.xlabel('Time')
    plt.ylabel('Power (W)')
    plt.title(f'Consumption vs. Time for Date: {date} {date.strftime("%A")}')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
total_power_per_date = grouped['Power (W)'].sum()

# Filter data for the last 5 dates (5 days)
last_5_dates = total_power_per_date.index[-5:]

# Create a week-long date range (5 days)
date_range = pd.date_range(start=last_5_dates[0], periods=5)

# Plot power consumption patterns for the last 5 dates over a week
plt.figure(figsize=(12, 6))
for date in last_5_dates:
    data = grouped.get_group(date)
    time_str = data['Time'].apply(lambda x: x.strftime('%H:%M:%S'))
    plt.plot(time_str, data['Power (W)'], label=f'{date.strftime("%A")}', marker='o')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
plt.gcf().autofmt_xdate()
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.title('Power Consumption Patterns for the Last 5 Dates Over a Week')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
X = df['People Inside'].values.reshape(-1, 1)  # Independent variable: Number of people
y = df['Power (W)'].values  # Dependent variable: Power consumption

# Step 2: Create a linear regression model
model = LinearRegression()

# Step 3: Fit the linear regression model
model.fit(X, y)

# Step 4: Make predictions
y_pred = model.predict(X)

# Step 5: Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression line')
plt.xlabel('Number of People')
plt.ylabel('Power Consumption (W)')
plt.title('Linear Regression: Number of People vs. Power Consumption')
plt.legend()
plt.grid(True)
plt.show()

# Display the regression coefficients
slope = model.coef_[0]
intercept = model.intercept_
print(f'Linear Regression Equation: Power Consumption = {slope:.2f} * Number of People + {intercept:.2f}')
```
